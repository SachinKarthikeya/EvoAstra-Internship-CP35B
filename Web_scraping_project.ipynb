```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time
from urllib.parse import urljoin
```

```python
options = Options()
options.add_argument("--headless")  
driver = webdriver.Chrome(options=options)
```

```python
base_url = "https://ackodrive.com"
listing_url = base_url + "/collection/maruti-suzuki-cars/"
driver.get(listing_url)
```

```python
for _ in range(3):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
```

```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html, "html.parser")

cars = []

cards = soup.select(".BuyCarCard_card__AsGXF") 
```

```python
for card in cards:
    name = card.select_one(".BuyCarCard_carName__SAJVh").get_text(strip=True) if card.select_one(".BuyCarCard_carName__SAJVh") else ""
    fuel_trans = card.select_one(".BuyCarCard_contentSpecifications__4ELYd").get_text(strip=True) if card.select_one(".BuyCarCard_contentSpecifications__4ELYd") else ""
    price = card.select_one(".BuyCarCard_priceRange__2q8tm").get_text(strip=True) if card.select_one(".BuyCarCard_priceRange__2q8tm") else ""
    year = card.select_one(".year").get_text(strip=True) if card.select_one(".year") else ""
    rate = card.select_one(".BuyCarCard_ratingVal__4zoYr").get_text(strip=True) if card.select_one(".BuyCarCard_ratingVal__4zoYr") else ""
    location = "Mumbai"  

    cars.append({
        "Name": name,
        "Fuel Type and Transmission": fuel_trans,
        "Price": price,
        "Year": year,
        "Rating": rate,
        "Location": location,
    })

driver.quit()

df = pd.DataFrame(cars)
df.to_csv("maruti_mumbai_ackodrive.csv", index=False)

print("CSV Saved!")
```
